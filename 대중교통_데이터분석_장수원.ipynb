#7시 최대승차량

```python
max=0
maxs=''
for row in data2:
  if row[10] >max:
    max=row[10]
    maxs=row[3]+row[1]
print(maxs, max)

result=[]
for row in data2:
  result.append(row[10])

print(result)
```
